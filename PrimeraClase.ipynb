```python
a = input("Introduce el valor del limite inferior")
b = input("Introduce el valor del limite superior")
n = input("Introduce el n√∫mero de iteraciones a resolver") 

a = int(a)
b = int(b)
n = int(n)
```

```python
from math import *
#funcion auxiliar para ejecutar f(x)
def fx(x, f):
    return eval(f)

f = '(x)'
f2 = '(x)**2'

def simpson(n, a, b, f):
    h = (b - a) / n
    print(h)
    suma = 0.0
    for i in range(1, n):       
        x = a + i * h
        print("valor x",i, "= ",x)
        if(i % 2 == 0):
            #si es par lo vamos a multiplicar por 2
            suma = suma + 2 * fx(x, f)
            print("valor suma par",i,"=",suma)
        else:
            #si es impar lo vamos a multiplicar por 4
            suma = suma + 4 * fx(x, f)
            print("valor suma impar",i,"=",suma)
            
    suma = suma + fx(a, f) + fx(b, f)
    resultado = suma * (h / 3)
    return (resultado)
print(simpson(n, a, b, f))
```
